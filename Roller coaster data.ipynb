## Step 4: Data wrangling

Includes processes such as removing irrelevant rows and column, dealing with missing data, identifying duplicate items and renaming columns.

```python
#to drop columns that are not relevant
df = df[['coaster_name', #'Length', 'Speed', 
         'Location', 
         'Status',# 'Opening date', 'Type', 
         'Manufacturer', #'Height restriction', 'Model', 'Height',
       #'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
       #'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
       #'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       #'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date', 'Opened', 'Replaced by', 'Website',
       #'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
       #'Single rider line available', 'Restraint Style',
       #'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', #'height_value', 'height_unit', 
         'height_ft',
       'Inversions_clean', 
         'Gforce_clean']].copy()
```

## Step 7: Feature relationships

Relation between different attributes
* Scatterplot
* Heatmap Correlation
* Pairplot
* GroupBy comparisons

```python
sp_he = df.plot(kind='scatter', x='Height_ft', 
        y='Speed_mph', color='green',
        title='Coaster Speed vs height')
sp_he.set_xlabel('Height (ft)')
sp_he.set_ylabel('Speed (mph)')

plt.show()
```

```python
sp_he1 = sns.scatterplot(x='Speed_mph', y='Height_ft', data=df, hue='Type_Main')
sp_he1.set(title='Coaster Speed vs height', xlabel='Speed (mph)', ylabel='Height (ft)')
plt.show()
```

```python
#to check correlation
df_corr = df[['Speed_mph', 'Height_ft', 'Gforce']].dropna().corr()
df_corr
```

```python
sns.heatmap(df_corr, annot=True, cmap= 'PuBuGn')
plt.show()
```

**Findings**
* Height in ft and Speed in mph have higher correlation of 0.73.  
* GForce and Height in ft have lower correlation of 0.47.

```python
#to get the locations with the fastest roller coasters 
ax = df.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean','count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean'] \
    .plot(kind='barh', figsize=(12, 5), title='Average Coast Speed by Location', color='indigo')
ax.set_xlabel('Average Coaster Speed')
plt.show()
```
